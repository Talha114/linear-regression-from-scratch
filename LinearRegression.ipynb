### Let's generate some random data

```python
X = np.arange(0,5,0.1, dtype=np.float32)
delta = np.random.uniform(-1,1, size=X.shape[0])
Y = .4 * X + 3 + delta

#making a copy for later use
rx = X
ry = Y
```

```python
plt.scatter(X,Y)
plt.xlabel('X')
plt.ylabel('Y')
```

### Implement functions for each of these

```python
theta1 = 0.0
theta2 = 0.0
m = X.shape[0]
```

$$
h(X, \theta) = X . \theta
$$

```python
def h(x,theta1,theta2):
    y = theta1 * x + theta2
    return y
```

$$
cost = \frac{1}{2m} \sum_{i = 0}^m{(h(X^i, \theta)-Y^i)}^2
$$

```python
def cost(m,theta1,theta2,X,Y):    
    t = 0
    for i in range(m):
        t += (((h(X[i],theta1,theta2) - Y[i]) ) ** 2)
    t /= 2*m
    return t
```

$$
\frac{\delta}{\delta \theta_j} = \frac{1}{m} \sum_{i = 0}^m{(h(X^i, \theta)-Y^i)} * X_j
$$

```python
def d1(m,theta1,theta2,X,Y):    
    t = 0
    for i in range(m):
        t += ((h(X[i],theta1,theta2) - Y[i]) ) * X[i]
    t /= m
    return t
```

```python
def d2(m,theta1,theta2,X,Y):   
    t = 0
    for i in range(m):
        t += ((h(X[i],theta1,theta2) - Y[i]) )
    t /= m
    return t
```

```python
def linearRegression(alpha,theta1,theta2):
    for i in range(500):
        der1 = d1(m,theta1,theta2,X,Y)
        der2 = d2(m,theta1,theta2,X,Y)
        theta1_new = theta1 - (der1 * alpha)
        theta2_new = theta2 - (der2 * alpha)
        theta1 = theta1_new
        theta2 = theta2_new
        print(cost(m,theta1,theta2,X,Y))
    return theta1,theta2
```

```python
linearRegression(0.2,theta1,theta2)
```

```python
t1,t2 = linearRegression(0.2,theta1,theta2)
```
